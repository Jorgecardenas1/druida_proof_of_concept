```python
import sys  
sys.path.insert(0, './Druida_POC/src/')


from druida import Stack
from druida import setup

from druida.DataManager import datamanager

import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn

from torchsummary import summary
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
from torcheval.metrics import BinaryAccuracy

random.seed(1)
```

### Architecture building

```python

#dnn.clear()
#dnn.pop()
#dnn.forward()
#dnn.layers

trainer = Stack.Trainer(learning_rate=0.001,batch_size=64, epochs=100)

dnn=Stack.DNN([{"name":"hidden1","layer":nn.Linear(8,50), "type":"hidden", "index":0},
               {"name":"activation1","layer":nn.ReLU(),"type":"activation", "index":1},
               {"name":"hidden2","layer":nn.Linear(50,8),"type":"hidden", "index":2},
               {"name":"activation2","layer":nn.ReLU(),"type":"activation", "index":3}, 
               {"name":"hidden3","layer":nn.Linear(8,1),"type":"hidden", "index":4},
               {"name":"output","layer":nn.Sigmoid(), "type":"output", "index":5}])

##checking device 
dnn.device
dnn.to(dnn.device)
```

### Input modeling

```python
#Data Loading
#In fact, we are supposed to have a fully functional dataset
#dataset=pd.read_csv('./Druida_POC/Data/prima-indians-diabetes.csv')
#dataset['Z'] = dataset['Z'].str.replace(',', '.')
#dataset['Z'] = dataset['Z'].astype(float)
#dataset.dtypes

#Y=dataset[['Z']]
#features = dataset[['h','X','Freq']]

dataset=pd.read_csv('./Druida_POC/Data/pima-indians-diabetes.csv')

train, test = train_test_split(dataset, test_size=0.2)

Y_train=train[['y']]
features_train = train[['x1','x2','x3','x4','x5','x6','x7','x8']]

Y_test=test[['y']]
feature_test = test[['x1','x2','x3','x4','x5','x6','x7','x8']]
```

```python


#Define the type of entry to use vector or image
trainX=datamanager.VectorSet(features_train)
trainY=datamanager.VectorSet(Y_train)

#do some moedling 
#it returns a tensor based on their vector or image type
#the goal is to implement solutions using either on or the ohter without
#having to do great changes in code.
X_train=trainX.modeling(
        {
            "type":setup.inputType['vector'],
            "size": (100,8),
            "torchType": torch.float32,
            "device":dnn.device
        }
)     

Y_train=trainY.modeling(
        {
            "type":setup.inputType['vector'],
            "size": (100,1),
            "torchType": torch.float32,
            "device":dnn.device
        }
)  


#lazzily load data
dataset_train = TensorDataset(X_train, Y_train)

#manage suffer, batching etc
trainLoader = DataLoader(dataset_train, batch_size=trainer.batch_size,shuffle=True)


#for x, y in trainLoader:
#    print(x)
#    print(y)
#    break

    
    
#Define the type of entry to use vector or image
testX=datamanager.VectorSet(feature_test)
testY=datamanager.VectorSet(Y_test)

#do some moedling 
#it returns a tensor based on their vector or image type
#the goal is to implement solutions using either on or the ohter without
#having to do great changes in code.
X_test=testX.modeling(
        {
            "type":setup.inputType['vector'],
            "size": (100,8),
            "torchType": torch.float32,
            "device":dnn.device
        }
)     

Y_test=testY.modeling(
        {
            "type":setup.inputType['vector'],
            "size": (100,1),
            "torchType": torch.float32,
            "device":dnn.device
        }
)  


#lazzily load data
dataset_test = TensorDataset(X_test, Y_test)

#manage suffer, batching etc
testLoader = DataLoader(dataset_test, batch_size=trainer.batch_size,shuffle=True)
```

```python

#optimizer
optimizer = torch.optim.SGD(dnn.parameters(), lr=trainer.learning_rate )
#loss function
loss_fn = nn.BCELoss()


```

### Training

```python
#Arrays to log and plot loss values



def train_loop(dataloader, model, loss_fn, optimizer, batch, acc,loss_val):
    size = len(dataloader)
    # Set the model to training mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    model.train()

    X,y=dataloader
    
    size = len(X)


    # Compute prediction and loss

    #prediction
    pred = model(X)
    loss = loss_fn(pred, y)
    optimizer.zero_grad()

    # Backpropagation
    loss.backward()
    optimizer.step()
        
    running_loss = loss.item()
    current =  (batch + 1) * len(X)
    #print(f"Training Loss: {loss:>7f}  [{current:>5d}/{size:>5d}] \n")
    acc = (pred.argmax(1) == y).type(torch.float).mean().item()

    print(f"Training: \n Accuracy: {100*acc}%, Loss: {loss:>7f}  [{current:>5d}/{size:>5d}] \n")
    
    return acc, running_loss

    



def test_loop(dataloader, model, loss_fn,batches, batch,acc_test,test_loss):
    # Set the model to evaluation mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    model.eval()
    num_batches = batches

    # Evaluating the model with torch.no_grad() ensures that no gradients are computed during test mode
    # also serves to reduce unnecessary gradient computations and memory usage for tensors with requires_grad=True
    with torch.no_grad():
        X, y = (dataloader)
        size = len(X)

        pred = model(X)
        test_loss = loss_fn(pred, y).item()
        acc_test = (pred.argmax(1) == y).type(torch.float).mean().item()
           
    
    #test_loss /= num_batches
    print(f"Test Error: \n Accuracy: {100*(acc_test):>0.1f}%, Avg loss: {test_loss:>8f} \n")
    
    return acc_test, test_loss
```

```python
loss_values,test_loss_values,train_acc_hist,test_acc=trainer.training(train_loop,test_loop, trainLoader,testLoader, dnn, loss_fn, optimizer)
```

```python
print('Training Accuracy='+str(np.mean(test_acc)))

from matplotlib import pyplot as plt
plt.plot(loss_values, label='train_loss')
plt.legend()
plt.show

```

```python
print('Testing Accuracy='+str(np.mean(train_acc_hist)))
plt.plot(train_acc_hist, label="train")
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```
